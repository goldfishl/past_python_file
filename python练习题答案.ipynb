```python
# 检查学生信息字段的规则
# 不符合规则：返回True  符合规则：返回False
def check_fields(stu_info):
    if list(stu_info.keys()) != ["姓名", "语文", "数学", "英语"]:
        print("学生信息中的字段必须为 ‘姓名’，‘语文’，‘数学’，‘英语’")
        return True
    elif type(stu_info["姓名"]) != str:
        print("姓名信息需要为字符串类型")
        return True
    elif type(stu_info["语文"]) != int or type(stu_info["数学"]) != int or type(stu_info["英语"]) != int:
        print("分数需要为整数类型")
        return True
    elif check_score(stu_info["语文"]) or check_score(stu_info["数学"]) or check_score(stu_info["英语"]):
        print("分数的范围必须在0-100之间")
        return True
    return False

# 检查分数是否在0-100的范围内
# 不在范围内：返回True  在范围内：返回False
def check_score(value):
    if value > 100 or value < 0:
        return True
    else:
        return False

# 检查学生是否存在
# 存在：返回True  不存在：返回False
def check_num(stu_num):
    class_names = ["一班", "二班"]
    for class_name in class_names:
        if stu_num in info[class_name].keys():
            return True
    return False

# 班级的判断
def judge(class_name):
    if class_name == '一班' or class_name == '二班':
        return False
    return True


def add_stu():
    stu_num = int(input('请输入学生学号：'))
    class_name = input('请输入班级名称：')
    stu_name = input('请输入学生姓名：')
    stu_chinese = int(input('请输入学生语文成绩：'))
    stu_math = int(input('请输入学生数学成绩：'))
    stu_eng = int(input('请输入学生英语成绩：'))
    stu_info = {'姓名': stu_name, '语文': stu_chinese, '数学': stu_math, '英语': stu_eng}

    stu_num = str(stu_num).zfill(3)
    if check_num(stu_num):
        print("学生已存在")
    else:
        if not (check_fields(stu_info) or judge(class_name)):
            info[class_name][stu_num] = stu_info
            print(info[class_name][stu_num]["姓名"]+"信息已添加")
        else:
            print("添加的班级或者数据不符合字段要求")
```

```python
def del_stu():
    stu_num = int(input('请输入学生学号：'))
    class_name = input('请输入班级名称：')
    stu_num = str(stu_num).zfill(3)
    if not check_num(stu_num):
        print("学生不存在")
    else:
        if not judge(class_name):
            print(info[class_name][stu_num]["姓名"]+"信息已删除")
            del info[class_name][stu_num]
        else:
            print('添加的班级或者数据不符合字段要求')
```

```python
def updata_stu():
    stu_num = int(input('请输入学生学号：'))
    class_name = input('请输入班级名称：')
    field = input('请输入需要修改的科目：')
    value = int(input('请输入需要更新的分数：'))
    fields = ["姓名", "语文", "数学", "英语"]
    stu_num = str(stu_num).zfill(3)
    if not judge(class_name):
        if stu_num in info[class_name].keys():
            if field in fields:
                if field == fields[0] and type(value) != str:
                    print("姓名信息需要为字符串类型")
                elif (field in fields[1:]) and type(value) != int:
                    print(field+"分数需要为整数类型")
                elif (field in fields[1:])  and check_score(value) :
                    print(field+"分数的范围必须在0-100之间")
                else:
                    name = info[class_name][stu_num]["姓名"]
                    info[class_name][stu_num][field] = value
                    print(name+"的"+field+"已更新为"+str(value))
            else:
                print("要进行更新的字段不存在")
        else:
            print("学生不在相应班级或学生不存在")
    else:
        print("班级输入错误")
```

```python
# judge函数不需要修改，修改会使程序出错
def judge1(c_name,s_name):
    if c_name == '一班' or c_name == '二班':
        if(s_name == '语文' or s_name == '数学' or s_name == '英文'):
            return False
    return True

def return_ave():
    class_name = input('请输入班级名称：')
    subject_name = input('请输入科目名称：')
    jud = judge1(class_name,subject_name) # 此处judge作为判断输入是否合法，judge为0时，程序正常运行，为1时提示输入错误
    
    if not jud:
        count = 0
        for i in list(info[class_name].keys()):
            count += info[class_name][i][subject_name]
        ave = count/len(info[class_name].keys())
        ave = class_name + '的' + subject_name + '考试成绩为' + str(ave) + '分'
    else:
        ave = '用户输入不合法，请重新选择功能'
    return print(ave)
```

```python
from pyecharts.charts import Bar

# chinese:存储遍历到的每个学生的语文成绩
# math:存储遍历到的每个学生的数学成绩
# eng:存储遍历到的每个学生的英语成绩
# class_list:分别存储每个班所有学生的总分
# scores_list:存储两个班总成绩的均值 的列表

def paint_ave():
    class_list = {'一班': 0, '二班':0}
    for i in list(info.keys()):
        for j in list(info[i].keys()):
            chinese = info[i][j]['语文']
            math = info[i][j]['数学']
            eng = info[i][j]['英语']
            class_list[i] += chinese + math + eng
        class_list[i] = class_list[i]/len(list(info[i].keys()))
        
    scores_list = list(class_list.values())
    bar = Bar()
    bar.add_xaxis(['一班','二班'])
    bar.add_yaxis("总分平均分", scores_list, bar_width='40')
    return bar.render('bar2.html')


# paint_ave()
```

```python
from IPython.display import clear_output as clear

def require():
    re = input('是否继续操作？（输入‘是’为继续操作，其他为退出系统）')
    if re == '是':
        clear()
    else:
        return 0

y = input('是否进入系统（输入‘是’为继续操作，其他为退出系统）：')
if y == '是':
    while True:
        print('1.添加一个学生的信息')
        print('2.删除一个学生的信息')
        print('3.更新一个学生的信息')
        print('4.返回指定班某科的平均值')
        print('5.输出两个班级三科总分的平均值柱状图')
        choose = input('请输入')
        if choose == '1':
            add_stu()
        elif choose == '2':
            del_stu()
        elif choose == '3':
            updata_stu()
        elif choose == '4':
            return_ave()
        elif choose == '5':
            paint_ave()
        else:
            print('输入错误，返回主界面')
        if(require() == 0):
                break

clear()
print('退出系统')
```
