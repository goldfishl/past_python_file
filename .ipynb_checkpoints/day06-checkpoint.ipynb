# Numpy
Numpy（Numerical Python）是一个开源的Python科学计算库，用于快速处理任意维度的数组。

## `array()`
它提供了常规Python列表list的替代品数据类型：numpy的ndarray（n维数组）。

ndarray（n维数组）数据类型一般是通过`array()`方法创建

```python
weight = [45, 65, 75]

print(type(weight))
```

```python
import numpy as np 

np_weight = np.array(weight)
print(type(np_weight))
```

## Numpy重点！！！！
list中可以存放不同的数据类型

但在numpy中ndarray里的值只能是<font color=red size=5>同一种数据类型！！！</font>

### 自动转换为同一种类型
如果你试图创建一个不同数据类型元素的ndarray,产生的ndarray将被转换为单一数据类型。

规则如下：
- 如果包含字符串类型元素则被转换为字符串类型：
    - 如果包含布尔、字符串和浮点数等数据类型的元素都被转换为字符串元素的ndarray

```python
weight = [45, 65, "long", False]

np_weight = np.array(weight)
print(np_weight)
```

- 不包含字符串类型元素时，布尔型数据类型元素会被转换为数值型数据类型元素
    - `True`转换为`1`，`False`转换为`0`

```python
weight = [45, 65, True]

np_weight = np.array(weight)
print(np_weight)
print(np_weight.dtype)
```

```python
weight = [45, 65, 56.5, True, False]

np_weight = np.array(weight)
print(np_weight)
print(np_weight.dtype)
```

- 当然也可以通过`np.array()`中的`dtype`参数来显示指定
    - 但要注意精度损失和可转换性

```python
weight = [45, 65, 56.5, True, False]

np_weight = np.array(weight, dtype=np.int64)
print(np_weight)
```

```python
weight = [45, 65, "long", False]

np_weight = np.array(weight, dtype=np.int64)
print(np_weight)
```

## ndarray的优势
### 内存存储及读写
出ndarray在存储数据的时候，数据与数据的地址都是连续的，这样就给使得批量操作数组元素时速度更快。

<div align=center>

</div>

这是因为ndarray中的所有元素的类型都是相同的，而Python列表中的元素类型是任意的，所以ndarray在存储元素时内存可以连续，而python 原生list就只能通过寻址方式找到下一个元素，这虽然也导致了在通用性能方面Numpy的ndarray不及Python原生list，但在科学计算中，Numpy 的ndarray就可以省掉很多循环语句，代码使用方面比Python原生list简单的多。

### ndarray支持并行化运算（向量化运算） 
numpy内置了并行运算功能，当系统有多个核心时，做某种计算时，numpy会自动做并行计算

### 效率远高于纯Python代码 
Numpy底层使用C语言编写，内部解除了GIL（全局解释器锁），其对数组的操作速度不受Python解释器的限制，所以，其效率远高于纯 Python代码。

### 代码简洁
可以直接对ndarray进行操作符运算。

##  ndarray与Python原生list运算效率对比

```python
import random 
import time 
import numpy as np 

a = [] 
for i in range(1000000):
    a.append(random.random()) # 通过%time魔法方法, 查看当前行的代码运行一次所花费的时间 

%time sum1=sum(a) 

b=np.array(a) 
%time sum2=np.sum(b)
```

# Numpy子集
## 索引、切片
ndarray有着和list一样的语法规则。

但在高维数组索引时，ndarray有着更简单的规则，我们会放在后面2维数组来讲。
## boolean布尔值索引
- 一般来说，首先使用比较运算符获得一个布尔类型的ndarray
    - 如果相应的元素符合比较条件则为真，不符合则为假。
- 然后在方括号内使用这个布尔数组来做“索引”，所以ndarray对应的布尔值是True的元素，就被选中组成新的ndarray。

```python
import numpy as np

bmi = np.array([18.49112426035503, 20.515086478979924, 22.89307408198773])

is_lightweight = bmi <21
print(is_lightweight)
```

```python
bmi[is_lightweight]
```

### Numpy中的布尔运算
numpy中布尔类型的ndarray的布尔运算是不能用经典python中的`and`、`or`、`not`的

```python
import numpy as np

bmi = np.array([18.49112426035503, 20.515086478979924, 22.89307408198773])

print(bmi < 22 and bmi > 21)
```

应该使用numpy中对应的`logical_and()`、`logical_or()`和`logical_not()`这三个函数

```python
print(np.logical_and(bmi < 22,bmi > 20))
```
