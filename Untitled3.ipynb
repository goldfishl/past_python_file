````python
from faker import Faker
import numpy as np
import random

# 随机生成分数
random.seed(123)
np.random.seed(123)
score = np.random.normal(loc =74.0 , scale= 10.0,size = (3,120))
score = score.astype(np.int)

# 随机生成姓名
name = set()
while len(name) != 120:
#     np.random.seed(123)
    Faker.seed(np.random.randint(low=0,high=3000))
    fake = Faker("zh_CN")
    name.add(fake.name())    
name = list(name)

# 创建学生信息
field = ["姓名", "语文", "数学", "英语"]
class_info = {"一班":{}, "二班":{}}
for i in range(120):
    val = [name[i], score[0][i], score[1][i], score[2][i]]
    stu_num = str(i+1).zfill(3)
    if i < 60:        
        class_info["一班"][stu_num] = dict(zip(field, val))
    else:
        class_info["二班"][stu_num] = dict(zip(field, val))
````

> ### 4、在指定班中添加一个学生的信息
>
> <font size=3>解析：使用函数在某个班里添加某个学生的信息</font>  
>
> <font size=3>要求：使用函数创建</font>  
>
> <font size=3>函数名：add_stu</font> 
>
> <font size=3>传入参数：stu_num：学号(整数)，class_name：班级名称(字符串)，学生信息（字典）</font>  
>
>
> <font size=3>传入参数示例：`add_stu(125, "二班", {'姓名': "王刚", '语文': 100, '数学': 65, '英语': 73})`</font>  
>
> <font size=3>返回值：无</font>  
>  
> **<font size=4>代码：</font>**  
> ```python
> def add_stu(stu_num, class_name, stu_info):
>     函数体
> ```

````python
# 检查学生信息字段的规则
# 不符合规则：返回True  符合规则：返回False
def check_fields(stu_info):
    if list(stu_info.keys()) != ["姓名", "语文", "数学", "英语"]:
        print("学生信息中的字段必须为 ‘姓名’，‘语文’，‘数学’，‘英语’")
        return True
    elif type(stu_info["姓名"]) != str:
        print("姓名信息需要为字符串类型")
        return True
    elif type(stu_info["语文"]) != int or type(stu_info["数学"]) != int or type(stu_info["英语"]) != int:
        print("分数需要为整数类型")
        return True
    elif check_score(stu_info["语文"]) or check_score(stu_info["数学"]) or check_score(stu_info["英语"]):
        print("分数的范围必须在0-100之间")
        return True
    return False

# 检查分数是否在0-100的范围内
# 不在范围内：返回True  在范围内：返回False
def check_score(value):
    if value > 100 or value < 0:
        return True
    else:
        return False

# 检查学生是否存在
# 存在：返回True  不存在：返回False
def check_num(stu_num):
    class_names = ["一班", "二班"]
    for class_name in class_names:
        if stu_num in class_info[class_name].keys():
            return True
    return False

# 
def add_stu(stu_num, class_name, stu_info):
    stu_num = str(stu_num).zfill(3)
    if check_num(stu_num):
        print("学生已存在")
    else:
        if not check_fields(stu_info):
            class_info[class_name][stu_num] = stu_info
            print(class_info[class_name][stu_num]["姓名"]+"信息已添加")
        else:
            print("添加的学数据不符合字段要求")
        

def del_stu(stu_num, class_name):
    stu_num = str(stu_num).zfill(3)
    if not check_num(stu_num):
        print("学生不存在")
    else:
        print(class_info[class_name][stu_num]["姓名"]+"信息已删除")
        del class_info[class_name][stu_num]

def updata_stu(stu_num, class_name, field, value):
    fields = ["姓名", "语文", "数学", "英语"]
    stu_num = str(stu_num).zfill(3)
    if stu_num in class_info[class_name].keys():
        if field in fields:
            if field == fields[0] and type(value) != str:
                print("姓名信息需要为字符串类型")
            elif (field in fields[1:]) and type(value) != int:
                print(field+"分数需要为整数类型")
            elif (field in fields[1:])  and check_score(value) :
                print(field+"分数的范围必须在0-100之间")
            else:
                name = class_info[class_name][stu_num]["姓名"]
                class_info[class_name][stu_num][field] = value
                print(name+"的"+field+"已更新为"+str(value))
        else:
            print("要进行更新的字段不存在")
    else:
        print("学生不在相应班级或学生不存在")
````
